```python
import sys
sys.path.append(r"C:\Projeto_SRAG")  # raiz do projeto

import duckdb
from config import SRAG_2023, SRAG_2024

con = duckdb.connect()

# Quantas linhas tem cada arquivo
con.sql(f"""
    SELECT '2023' AS ano, COUNT(*) AS total
    FROM read_csv('{SRAG_2023}', delim=';', header=true)
    UNION ALL
    SELECT '2024', COUNT(*)
    FROM read_csv('{SRAG_2024}', delim=';', header=true)
""").show()

# Ver as primeiras linhas
con.sql(f"""
    SELECT *
    FROM read_csv('{SRAG_2023}', delim=';', header=true)
    LIMIT 5
""").show()

# Quantas colunas e quais são
con.sql(f"""
    DESCRIBE
    SELECT *
    FROM read_csv('{SRAG_2023}', delim=';', header=true)
""").show()
```

```python
import sys
sys.path.append(r"C:\Projeto_SRAG")

import duckdb
from config import SRAG_2023, SRAG_2024

con = duckdb.connect()

con.sql(f"""
    SELECT
        COD_IDADE,
        COUNT(*) AS total
    FROM read_csv(
        '{SRAG_2023}',
        delim=';',
        header=true,
        types={{'COD_IDADE': 'VARCHAR'}}
    )
    GROUP BY COD_IDADE

    UNION ALL

    SELECT
        COD_IDADE,
        COUNT(*) AS total
    FROM read_csv(
        '{SRAG_2024}',
        delim=';',
        header=true,
        types={{'COD_IDADE': 'VARCHAR'}}
    )
    GROUP BY COD_IDADE

    ORDER BY total DESC
    LIMIT 30
""").show()
```

```python
import duckdb
import sys
sys.path.append(r"C:\Projeto_SRAG")
from config import SRAG_SILVER

con = duckdb.connect()

# Shape geral
con.sql(f"""
    SELECT COUNT(*) AS total_registros
    FROM '{SRAG_SILVER}'
""").show()

# Distribuição por ano
con.sql(f"""
    SELECT ANO_NOTIFIC, COUNT(*) AS total
    FROM '{SRAG_SILVER}'
    GROUP BY ANO_NOTIFIC
""").show()

# Checagem de evolução
con.sql(f"""
    SELECT EVOLUCAO, COUNT(*) AS total
    FROM '{SRAG_SILVER}'
    GROUP BY EVOLUCAO
    ORDER BY total DESC
""").show()

# Checagem de idades absurdas
con.sql(f"""
    SELECT 
        MIN(IDADE_ANOS) AS menor_idade,
        MAX(IDADE_ANOS) AS maior_idade,
        AVG(IDADE_ANOS) AS media_idade
    FROM '{SRAG_SILVER}'
""").show()
```
